```python
import os
import numpy as np
import json
import pandas as pd
```

```python
data = {}
for file in os.listdir('results'):
    if file[-5:] == '.json':
        with open(f'results/{file}') as json_file:
            dataset_dict = json.load(json_file)
            print(dataset_dict)
            for key, value in dataset_dict.items():
                data[key] = value
    print(file)
```

```python
data.keys()
```

```python
data['ogbg-molbace']['results'].keys()
```

```python
df = []

for dataset_name, dataset_data in data.items():
    l = [
        dataset_name,
        dataset_data['metric'],
        dataset_data['runs'],
    ]

    for subset in ['train','valid','test']:
        mean = np.mean(dataset_data['results']['path2vec'][subset])
        std = np.std(dataset_data['results']['path2vec'][subset])
        l.append(f'{mean:.4f} +- {std:.4f}')

    mean = np.mean(dataset_data['results']['random'][subset])
    std = np.std(dataset_data['results']['random'][subset])
    l.append(f'{mean:.4f} +- {std:.4f}')

    df.append(l)

df = pd.DataFrame(df, columns=['dataset', 'metric', 'runs', 'train','valid','test', 'test random'])
df
```

```python
classification_df= df[df['metric'] != 'rmse']
classification_df = classification_df.drop(labels=['test random'], axis=1)
classification_df
```

```python
regression_df= df[df['metric'] == 'rmse']
regression_df
```
